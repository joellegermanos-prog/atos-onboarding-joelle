```python
# Cellule de setup — lance-la, et c'est tout
import sys
print(f"Python : {sys.version.split()[0]}")

try:
    import pandas as pd
    print(f"pandas : {pd.__version__}")
except ImportError:
    print("⚠️  pandas n'est pas installé. Lance : pip install pandas")

try:
    import numpy as np
    print(f"numpy : {np.__version__}")
except ImportError:
    print("⚠️  numpy n'est pas installé. Lance : pip install numpy")

# Compteur de score (à laisser tel quel)
SCORE = {"total": 0, "reussi": 0}
```

```python
# Complète cette fonction
def pairs_au_carre(liste):
    # Ton code ici
    return [x**2 for x in liste if x % 2 == 0]
    pass


# --- Test (ne pas modifier) ---
SCORE["total"] += 1
try:
    assert pairs_au_carre([1, 2, 3, 4, 5]) == [4, 16]
    assert pairs_au_carre([0, -2, 7, 8]) == [0, 4, 64]
    assert pairs_au_carre([]) == []
    print("✅ Question 1 OK")
    SCORE["reussi"] += 1
except (AssertionError, TypeError) as e:
    print(f"❌ Question 1 — réessaie ({type(e).__name__})")
```

```python
# Complète cette fonction
def total_credits(transactions):
    # Ton code ici
    total = 0
    for t in transactions:
        if t["type"] == "credit":
            total += t["montant"]
    return total

# --- Test ---
SCORE["total"] += 1
try:
    txs = [
        {"id": 1, "montant": 100, "type": "credit"},
        {"id": 2, "montant": 50, "type": "debit"},
        {"id": 3, "montant": 200, "type": "credit"},
    ]
    assert total_credits(txs) == 300
    assert total_credits([]) == 0
    assert total_credits([{"id": 1, "montant": 99, "type": "debit"}]) == 0
    print("✅ Question 2 OK")
    SCORE["reussi"] += 1
except (AssertionError, TypeError) as e:
    print(f"❌ Question 2 — réessaie ({type(e).__name__})")
```

```python
# Complète cette fonction
def division_safe(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Division impossible"
    except TypeError:
        return "Type invalide"

# --- Test ---
SCORE["total"] += 1
try:
    assert division_safe(10, 2) == 5
    assert division_safe(7, 0) == "Division impossible"
    assert division_safe("dix", 2) == "Type invalide"
    print("✅ Question 3 OK")
    SCORE["reussi"] += 1
except (AssertionError, TypeError) as e:
    print(f"❌ Question 3 — réessaie ({type(e).__name__})")
```

```python
# Complète cette fonction
def compter_lignes(chemin):
    # Ton code ici
    with open(chemin, "r", encoding="utf-8") as f:
        # f.readlines() crée une liste où chaque élément est une ligne du fichier
        return len(f.readlines())
    pass


# --- Test ---
SCORE["total"] += 1
try:
    assert compter_lignes("transactions.csv") == 5
    print("✅ Question 4 OK")
    SCORE["reussi"] += 1
except (AssertionError, TypeError, FileNotFoundError) as e:
    print(f"❌ Question 4 — réessaie ({type(e).__name__})")
```

```python
import numpy as np

notes = np.array([12, 15, 8, 18, 11, 14, 9, 16, 13, 10])

# Ton code ici
moyenne = np.mean(notes)
ecart_type = np.std(notes)
mediane = np.median(notes)


# --- Test ---
SCORE["total"] += 1
try:
    assert abs(moyenne - 12.6) < 0.01
    assert abs(ecart_type - 3.04) < 0.05
    assert abs(mediane - 12.5) < 0.01
    print("✅ Question 7 OK")
    SCORE["reussi"] += 1
except (AssertionError, TypeError) as e:
    print(f"❌ Question 7 — réessaie ({type(e).__name__})")
```

```python
import pandas as pd

data = {
    "ville": ["Paris", "Lyon", "Marseille", "Toulouse", "Nice"],
    "population": [2150000, 515000, 870000, 480000, 340000],
    "region": ["Île-de-France", "Auvergne-Rhône-Alpes", "PACA", "Occitanie", "PACA"]
}

# Ton code ici
df = pd.DataFrame(data)


# --- Test ---
SCORE["total"] += 1
try:
    assert isinstance(df, pd.DataFrame)
    assert df.shape == (5, 3)
    assert list(df.columns) == ["ville", "population", "region"]
    print("✅ Question 8 OK")
    SCORE["reussi"] += 1
except (AssertionError, TypeError) as e:
    print(f"❌ Question 8 — réessaie ({type(e).__name__})")
```

```python
# Ton code ici (on suppose que df de la Q8 existe)
villes_grandes = df[df["population"] > 500000]

# --- Test ---
SCORE["total"] += 1
try:
    assert isinstance(villes_grandes, pd.DataFrame)
    assert len(villes_grandes) == 3
    assert set(villes_grandes["ville"]) == {"Paris", "Lyon", "Marseille"}
    print("✅ Question 9 OK")
    SCORE["reussi"] += 1
except (AssertionError, TypeError, NameError) as e:
    print(f"❌ Question 9 — réessaie ({type(e).__name__})")
```
