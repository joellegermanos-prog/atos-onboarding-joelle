```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Charge un dataset public (au choix). Ex Palmer Penguins :
URL = "https://raw.githubusercontent.com/allisonhorst/palmerpenguins/main/inst/extdata/penguins.csv"
df = pd.read_csv(URL)
df.head()
```

```python
df.describe(include="all")
```

```python
# Un graphique simple commenté
# Ex pour Palmer Penguins : masse moyenne par espèce
df.groupby("species")["body_mass_g"].mean().plot.bar(
    title="Masse moyenne par espèce — Palmer Penguins"
)
plt.ylabel("Masse corporelle (g)")
plt.show()
```

**Observation (1 ligne)** : *[ce que tu vois sur le graphique]*
Ce diagramme en barres démontre clairement que l'espèce Gentoo est la plus lourde avec une masse moyenne supérieure à 5000g , tandis que les espèces Adelie et Chinstrap ont un gabarit similaire aux alentours de 3700 g.
